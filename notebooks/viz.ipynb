```r
# Load required libraries
library(fmcsR)
library(igraph)
library(ChemmineR)
library(ChemmineOB)
smiles_to_inchikey <- function(smiles) {
    # Convert SMILES to mol object
    inchikey <-trimws(convertFormat("SMILES","INCHIKEY",smiles))
    return(inchikey)
}
```
